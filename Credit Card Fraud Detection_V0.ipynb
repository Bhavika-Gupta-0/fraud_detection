```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data and reading it through pandas
df= pd.read_csv("creditcard.csv")
df.head()
```

```python
# distribution of legit transactions & fraudulent transactions
df['Class'].value_counts()
```

#### This is highly unbalanced data set
0 - represents normal transaction
1 - represents fraudulent transaction

```python
# seperating the data for analysis

legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
df.groupby('Class')['Amount'].mean()
```

```python
# Under-Sampling
```

```python
# Built a sample dataset containing similar distribution of normal transactions and fraudulent tranactions

# Number of fraudulent transaction --> 492
```

```python
legit_sample = legit.sample(n=492)
legit_sample
```

```python
# concatenating two datasets

new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.groupby('Class').mean()
```

```python
# Splitting the data into features and target
```

```python
X= new_df.drop(columns = 'Class', axis=1)
Y= new_df['Class']
X
```

```python
Y
```

```python
# Split the data into training data & Testing data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##### Model Training

#### Logistic Regression 

```python
model = LogisticRegression()
```

```python
# Training the logistic Regression model with training data

model.fit(X_train, Y_train)
```

### Model Evaluation

####  Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy of training data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of test data: ", test_data_accuracy)
```
